# Simple Classification Using Perceptron Neuron

#### Implement a perceptron neuron model and train it using the AND gate dataset.

```python
import numpy as np
# make a prediction with a perceptron neuron
from sklearn.linear_model import Perceptron

# define dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 0, 0, 1]) 

# define perceptron model
net = Perceptron()
print(net)
```

```python
# fit (train) model
net.fit(X, Y)
```

```python
# Measure the accuracy of the model and displat it
accuracy = net.score(X, Y)
print(accuracy)
```

```python
net.coef_
```

```python
# define an input 
new_input = [1, 1]

# make a prediction for the given input
prediction = net.predict([new_input])
print(f"Prediction for input {new_input}: {prediction[0]}")
```

```python
# define an input 
new_input = [0.1, 0.2]

# make a prediction for the given input
prediction = net.predict([new_input])
print(f"Prediction for input {new_input}: {prediction[0]}")
```

```python
# define an input 
new_input = [1.2, 0.9]

# make a prediction for the given input
prediction = net.predict([new_input])
print(f"Prediction for input {new_input}: {prediction[0]}")
```

```python
# Display the learned weights and biases
print("Learned weights:", net.coef_)
print("Learned bias:", net.intercept_)
```
